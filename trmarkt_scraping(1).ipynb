```python
from bs4 import BeautifulSoup
import requests
import csv
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0"}
```

```python
player_name = []
player_values=[]
player_ages=[]
player_pos=[]
player_team=[]

x = 1
while x<=10:
    request = requests.get("https://www.transfermarkt.com.tr/spieler-statistik/wertvollstespieler/marktwertetop?land_id=0&ausrichtung=alle&spielerposition_id=alle&altersklasse=alle&jahrgang=0&kontinent_id=0&plus=3&page="+str(x)+"",headers=header)
    soup = BeautifulSoup(request.content,"lxml")


    pg = soup.find("div",{"class":"grid-view"})
    pg2 = pg.find("ul",{"class": "yiiPager"})
    pg3 = pg2.find_all("li",{"class":"page selected"})




    for content in pg3:
        a = content.a.get("href")
        b = "https://www.transfermarkt.com"
        link = b+a
        print(link)

        request2 = requests.get(link,headers=headers)
        soup2 = BeautifulSoup(request2.content,"lxml")

        players_name=soup2.find_all("a",{"class":"spielprofil_tooltip"})
        #print(players_name)    
        for player in players_name:
            #print(player.text)
            player_name.append(player.text)
        
        players_values=soup2.find_all("td",{"class":"rechts hauptlink"})
        for value in players_values:
            #print(value.text)
            player_values.append(value.text)
        
        players_ages=soup2.find_all("td",{"class":"zentriert"})
        for age in players_ages:
            #print(age.text)
            player_ages.append(age.text)
    
        positions=soup2.find_all("td",{"class":""})
        for position in positions:
            #print(position.text)
            player_pos.append(position.text)
            
        teams=soup2.find_all("img",{"class":""},alt=re.compile("^(?!http).*"))
        for team in teams:
            #print(team.get("alt"))
            player_team.append(team.get("alt"))
        

    x = x+1
```
